```python
import numpy as np
import matplotlib.pyplot as plt

# Предположим, что `samples` — это список списков, где каждый вложенный список — это выборка
# Например: samples = [[...], [...], ...]
# sizes — это список объемов выборок, например: sizes = [10, 30, 50, 100]

# Создадим словарь для хранения оценок по объему выборки
results = {size: {"means": [], "variances": []} for size in sizes}

# Заполним словарь результатами
for size, sample in zip(sizes, samples):
    mean = np.mean(sample)  # Оценка математического ожидания
    variance = np.var(sample, ddof=1)  # Оценка дисперсии (ddof=1 для несмещённой оценки)
    
    results[size]["means"].append(mean)
    results[size]["variances"].append(variance)

# Вычислим математическое ожидание и дисперсию для каждой группы
summary = {}
for size, data in results.items():
    summary[size] = {
        "mean_of_means": np.mean(data["means"]),  # Математическое ожидание оценок
        "variance_of_means": np.var(data["means"], ddof=1),  # Дисперсия оценок мат. ожидания
        "mean_of_variances": np.mean(data["variances"]),  # Математическое ожидание оценок дисперсии
        "variance_of_variances": np.var(data["variances"], ddof=1),  # Дисперсия оценок дисперсии
    }

# Выведем результаты
for size, data in summary.items():
    print(f"Объем выборки: {size}")
    print(f"  Математическое ожидание оценок: {data['mean_of_means']:.4f}")
    print(f"  Дисперсия оценок мат. ожидания: {data['variance_of_means']:.4f}")
    print(f"  Математическое ожидание оценок дисперсии: {data['mean_of_variances']:.4f}")
    print(f"  Дисперсия оценок дисперсии: {data['variance_of_variances']:.4f}")
    print()

# Построим графики
sizes_list = sorted(summary.keys())
mean_of_means = [summary[size]["mean_of_means"] for size in sizes_list]
variance_of_means = [summary[size]["variance_of_means"] for size in sizes_list]
mean_of_variances = [summary[size]["mean_of_variances"] for size in sizes_list]
variance_of_variances = [summary[size]["variance_of_variances"] for size in sizes_list]

# График математического ожидания оценок
plt.figure(figsize=(10, 5))
plt.plot(sizes_list, mean_of_means, marker='o', label="Математическое ожидание оценок")
plt.xlabel("Объем выборки")
plt.ylabel("Математическое ожидание")
plt.title("Зависимость математического ожидания оценок от объема выборки")
plt.grid(True)
plt.legend()
plt.show()

# График дисперсии оценок
plt.figure(figsize=(10, 5))
plt.plot(sizes_list, variance_of_means, marker='o', label="Дисперсия оценок мат. ожидания")
plt.xlabel("Объем выборки")
plt.ylabel("Дисперсия")
plt.title("Зависимость дисперсии оценок от объема выборки")
plt.grid(True)
plt.legend()
plt.show()

# График математического ожидания оценок дисперсии
plt.figure(figsize=(10, 5))
plt.plot(sizes_list, mean_of_variances, marker='o', label="Математическое ожидание оценок дисперсии")
plt.xlabel("Объем выборки")
plt.ylabel("Математическое ожидание")
plt.title("Зависимость математического ожидания оценок дисперсии от объема выборки")
plt.grid(True)
plt.legend()
plt.show()

# График дисперсии оценок дисперсии
plt.figure(figsize=(10, 5))
plt.plot(sizes_list, variance_of_variances, marker='o', label="Дисперсия оценок дисперсии")
plt.xlabel("Объем выборки")
plt.ylabel("Дисперсия")
plt.title("Зависимость дисперсии оценок дисперсии от объема выборки")
plt.grid(True)
plt.legend()
plt.show()
```
